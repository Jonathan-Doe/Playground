```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('data.csv')
X = data[['x1', 'x2']].values.reshape(-1,2)
x_max = max(X.T[0])
y = data.target.values
color = ['blue' if x == 1 else 'red' for x in y]
```

```python
def predict(W,X,b):
    return (np.matmul(X,W) + b >= 0).astype(int)

def perceptronstep(X, y, W, b, learning_rate = 0.01):
    
    for i in range(X.shape[0]): 
        y_hat = predict(W,X[i],b)
        
        if (y_hat == 1) & (y[i] == 0 ):
            W[0] -= learning_rate * X[i][0]
            W[1] -= learning_rate * X[i][1]
            b -= learning_rate 
        
        elif (y_hat == 0) & (y[i] == 1):
            W[0] += learning_rate * X[i][0]
            W[1] += learning_rate * X[i][1]
            b += learning_rate
    return W,b

def fit(X,y,learning_rate= 0.01, num_epoch = 25):
    x_max = max(X.T[0])
    W = np.random.rand(2,1)
    b = np.random.rand(1)[0] + x_max
    for i in range(num_epoch):
        W, b = perceptronstep(X,y,W,b,learning_rate=learning_rate)
    return W,b
```

```python
W_,b_ = fit(X,y)
```

```python
x1 = np.linspace(0,1,50)
x2 = (-b_ - W_[0]* x1) / W_[1]
```

```python
plt.figure(figsize = (12,9))
plt.title('Boundary line')
plt.scatter(X[:,0], X[:,1], color = color)
plt.plot(x1,x2, color = 'orange')
```
